```python
# FP
class treeNode:
    def __init__(self, nameValue, numOccur, parentNode):
        self.name = nameValue  #
        self.count = numOccur   # 
        self.nodeLink = None   # None
        self.parent = parentNode   # 
        self.children = {}  # 

    # 
    def inc(self, numOccur):
        self.count += numOccur

    # FP
    def disp(self, ind=1):
        print(' ' * ind, self.name, ' ', self.count)
        for child in self.children.values():
            child.disp(ind + 1)

# =======================================================FP==================================================

# 
def updateHeader(nodeToTest, targetNode):
    while (nodeToTest.nodeLink != None):
        nodeToTest = nodeToTest.nodeLink
    nodeToTest.nodeLink = targetNode

# FP
def updateTree(items, inTree, headerTable, count):
    if items[0] in inTree.children:
        # +1
        inTree.children[items[0]].inc(count)
    else:
        # 
        inTree.children[items[0]] = treeNode(items[0], count, inTree)
        # 
        if headerTable[items[0]][1] == None:  # 
            headerTable[items[0]][1] = inTree.children[items[0]]
        else:
            updateHeader(headerTable[items[0]][1], inTree.children[items[0]])

    if len(items) > 1:
        # updateTree
        updateTree(items[1::], inTree.children[items[0]], headerTable, count)

# FPdataSetminSup
def createTree(dataSet, minSup=1):
    # 
    headerTable = {}
    for trans in dataSet:
        for item in trans:
            headerTable[item] = headerTable.get(item, 0) + dataSet[trans]
    # 
    keys = list(headerTable.keys()) # 
    for k in keys:
        if headerTable[k] < minSup:
            del(headerTable[k])
    # 
    freqItemSet = set(headerTable.keys())
    if len(freqItemSet) == 0:
        return None, None
    # 
    for k in headerTable:
        headerTable[k] = [headerTable[k], None]  # 
    retTree = treeNode('Null Set', 1, None) # 
    # FP
    for tranSet, count in dataSet.items():
        localD = {} # 1
        for item in tranSet:
            if item in freqItemSet:   # 
                localD[item] = headerTable[item][0] # [0]
        if len(localD) > 0:
            orderedItems = [v[0] for v in sorted(localD.items(), key=lambda p: p[1], reverse=True)] # 
            updateTree(orderedItems, retTree, headerTable, count) # FP
    return retTree, headerTable

# ================================================================================================

# prefixPathleafNodeprefixPath
# prefixPathtreeNodetreeNode
def ascendTree(leafNode, prefixPath):
    if leafNode.parent != None:
        prefixPath.append(leafNode.name)
        ascendTree(leafNode.parent, prefixPath)

# basePettreeNodeFP
# condPats
def findPrefixPath(basePat, treeNode):
    condPats = {}  # 
    while treeNode != None:
        prefixPath = []  # 
        ascendTree(treeNode, prefixPath)  # 
        if len(prefixPath) > 1:
            condPats[frozenset(prefixPath[1:])] = treeNode.count  # 
        treeNode = treeNode.nodeLink  # 
    return condPats

# ================================================================================================
# FP
# FPFP
# FP

# inTreeheaderTablecreateTree()FP
# minSup
# preFixset([])
# freqItemList[]
def mineTree(inTree, headerTable, minSup, preFix, freqItemList):
    # 
    sorted_headerTable = sorted(headerTable.items(), key=lambda p: p[1][0])  #[key,[num,treeNode],()
    bigL = [v[0] for v in sorted_headerTable]  # 
    for basePat in bigL:
        newFreqSet = preFix.copy()  # 
        newFreqSet.add(basePat)     # 
        freqItemList.append(newFreqSet)  # 
        condPattBases = findPrefixPath(basePat, headerTable[basePat][1])  # basePat
        myCondTree, myHead = createTree(condPattBases, minSup)  # FP

        if myHead != None:
            # 
            print('conditional tree for:', newFreqSet)
            myCondTree.disp()
            mineTree(myCondTree, myHead, minSup, newFreqSet, freqItemList)  # 

# 
def loadSimpDat():
    simpDat = data
    return simpDat

# 
def createInitSet(dataSet):
    retDict = {}
    for trans in dataSet:
        retDict[frozenset(trans)] = 1
    return retDict

if __name__=='__main__':
    minSup = 0.1
    simpDat = loadSimpDat()  # 
    initSet = createInitSet(simpDat)  # 
    myFPtree, myHeaderTab = createTree(initSet, minSup)  # FP
    # myFPtree.disp()  # 

    freqItems = []  # 
    mineTree(myFPtree, myHeaderTab, minSup, set([]), freqItems)  # 
    print(freqItems)  # 
```

```python

for j in range(0,940):
    for i in range(0,8):
        if data[j][i] == 'freshmeat':
            data[j][i]=1
        elif data[j][i] == 'dairy':
            data[j][i]=2
        elif data[j][i] == 'confectionery':
            data[j][i]=3
        elif data[j][i] == 'cannedveg':
            data[j][i]=4
        elif data[j][i] == 'frozenmeal':
            data[j][i]=5
        elif data[j][i] == 'beer':
            data[j][i]=6
        elif data[j][i] == 'fruitveg':
            data[j][i]=7
        elif data[j][i] == 'wine':
            data[j][i]=8
        elif data[j][i] == 'softdrink':
            data[j][i]=9
        elif data[j][i] == 'cannedmeat':
            data[j][i]=10
        elif data[j][i] == 'fish':
            data[j][i]=11
        elif data[j][i] == 'nan':
            data[j][i].drop()
print(data)
#Apriori

def loadDataSet():
    return data

# 1dataSetlistset
def createC1(dataSet):
    C1 = []   # 1
    for transaction in dataSet:
        for item in transaction:
            if not [item] in C1:
                C1.append([item])
    C1.sort()  # n
    # 1n1
    return list(map(frozenset, C1))   #map(frozenset, C1)C1PythonfrozensetPython

# 
# dataSetCkkkminSupport
def scanD(dataSet, Ck, minSupport):
    ssCnt = {}   # 
    for tid in dataSet:
        for can in Ck:
            if can.issubset(tid):
                ssCnt[can] = ssCnt.get(can, 0) + 1   # 
    numItems = float(len(dataSet))
    retList = []
    supportData = {}
    for key in ssCnt:
        support = ssCnt[key] / numItems
        if support >= minSupport:
            retList.insert(0, key)  #
        supportData[key] = support
    return retList, supportData   #retListCkminSupportsupportData

# LkkC(k+1)
def aprioriGen(Lk, k):
    retList = []
    lenLk = len(Lk)
    for i in range(lenLk):
        for j in range(i + 1, lenLk):
            # k-1k+1
            L1 = list(Lk[i])[:k-2]; L2 = list(Lk[j])[:k-2]   # k-1
            L1.sort(); L2.sort()
            if L1 == L2:
                retList.append(Lk[i] | Lk[j])
    return retList

# 
# CkkC1createC1()LkksupKLksupKscanD()Ck
def apriori(dataSet, minSupport=0.2):
    C1 = createC1(dataSet)  # 1
    D = list(map(set, dataSet))  # 
    L1, supportData = scanD(D, C1, minSupport)  # 1
    L = [L1]  # L
    k = 2
    while (len(L[k-2]) > 0): # n
        Ck = aprioriGen(L[k-2], k)   # Ckk
        Lk, supK = scanD(D, Ck, minSupport)  # LkksupK
        L.append(Lk);supportData.update(supK)  # 
        k += 1
    return L, supportData

if __name__=='__main__':
    dataSet = loadDataSet()  # 
    # C1 = createC1(dataSet)  # 1
    # D = list(map(set, dataSet))  # 
    # L1, suppDat = scanD(D, C1, 0.5)
    # print(L1,suppDat)

    L, suppData = apriori(dataSet,minSupport=0.1)
    print(L,suppData)
```

```python
def calcConf( freqSet, H, supportData, brl, minConf=0.8 ):
    '''
    
    
    freqSet(frozenset):
    H(frozenset):
    supportData(dic):
    brl(tuple):
    minConf(float):
    '''
    prunedH = []
    for conseq in H:
        conf = supportData[ freqSet ] / supportData[ freqSet - conseq ]
        if conf >= minConf:
            print (freqSet - conseq, '-->', conseq, 'conf:', conf)
            brl.append( ( freqSet - conseq, conseq, conf ) )
            prunedH.append( conseq )
    return prunedH

def rulesFromConseq( freqSet, H, supportData, brl, minConf=0.8 ):
    '''
    2
    
    freqSet(frozenset):
    H(frozenset):
    supportData(dict):
    brl(tuple):
    
    '''
    m = len( H[ 0 ] )
    
    #  m
    if len( freqSet ) > m + 1:
        Hmp1 = aprioriGen( H, m + 1 )
        Hmp1 = calcConf( freqSet, Hmp1, supportData, brl, minConf )
        
        # 
        if len( Hmp1 ) > 1:
            rulesFromConseq( freqSet, Hmp1, supportData, brl, minConf )

def generateRules( L, supportData, minConf=0.8 ):
    '''
    
    
    L(list):
    supportData(dict):
    minConf(float):
    '''
    bigRuleList = []
    for i in range( 1, len( L ) ):
        for freqSet in L[ i ]:
            # freqSet
            H1 = [ frozenset( [ item ] ) for item in freqSet ]
            
            # 2
            if i > 1:
                rulesFromConseq( freqSet, H1, supportData, bigRuleList, minConf )
            else:
                calcConf( freqSet, H1, supportData, bigRuleList, minConf )
    return bigRuleList

if __name__ == '__main__':
    # 
    myDat = loadDataSet()    
    # 
    L, suppData = apriori( myDat, 0.1 )
    
    rules = generateRules( L, suppData, minConf=0.8 )
    print( 'rules:\n', rules)
```
