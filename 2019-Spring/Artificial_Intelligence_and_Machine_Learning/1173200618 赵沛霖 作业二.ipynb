基于存在交互项多元线性回归模型预测学生被录取的概率

一、分析查看数据

```python
import pandas as pd

df = pd.read_csv('Admission_Predict_Ver1.1.csv',index_col=0)
df.shape
```

```python
df.head(5)
```

```python
X = df.iloc[:,:7]
X.sample(5)
```

```python
y=df.iloc[:,-1]
y.sample(5)
```

二、进行线性回归预测

```python
from sklearn.model_selection import train_test_split

X_tr, X_ts, y_tr, y_ts = train_test_split(X, y)
X_tr.shape, X_ts.shape
```

```python
import statsmodels.formula.api as smf
model = smf.OLS(y, X)
model_result = model.fit()
```

```python
model_result.summary()
```

分析：模型拟合程度非常好Prob (F-statistic)接近于0。在这几个自变量中，SOP的P值大于0.05，因此需要把该项去掉，同时可能存在交互项，需要排除交互项的相互干扰。

```python
tmp=df['University Rating'].copy()
df['UR']=tmp

del tmp
```

```python
tmp=df['TOEFL Score'].copy()
df['TS']=tmp

del tmp
```

```python
tmp=df['GRE Score'].copy()
df['GS']=tmp

del tmp
```

```python
tmp=df['LOR '].copy()
df['LOR']=tmp

del tmp
```

```python
tmp=df['Chance of Admit '].copy()
df['COA']=tmp

del tmp
```

```python
df.head(5)
```

```python
import statsmodels.formula.api as smf
model = smf.ols('COA~GS+TS+LOR+CGPA+Research+UR+SOP+UR:SOP', df).fit()
model.summary()
```

分析：University Rating和Statement of Purpose and Letter of Recommendation Strength存在交互影响，在排除影响之后两者P值均小于0.05，通过检验。

三、进行预测

```python
y_hat = model_result.predict(X)
y_hat = pd.DataFrame(y_hat, columns=['pred'])
df_merge = df.merge(y_hat, left_index=True, right_index=True)
df_merge.head()
```

```python
import numpy as np
y = df['COA']
y_hat=df_merge['pred']
model_RMSE = np.sqrt(np.mean(np.square(y_hat-y)))
model_RMSE
```

总结：此次作业运用存在交互项的多元线性回归模型进行预测。均方误差MSE很小，同时模型的R值为0.825，说明模型精度很高。模型的P值非常小，通过假设检验，模型通过检验；同时，每一个自变量的P值都非常小，都通过检验，而且排除了University Rating和SOP的相互影响。
