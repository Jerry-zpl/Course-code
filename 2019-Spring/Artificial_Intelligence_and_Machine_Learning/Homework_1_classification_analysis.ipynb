三、将变量转化为数值

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_date = le.fit(df['Date'])
df['Date'] = df_date.transform(df['Date'])

df_location = le.fit(df['Location'])
df['Location'] = df_location.transform(df['Location'])

df_raintomorrow = le.fit(df['RainTomorrow'])
df['RainTomorrow'] = df_raintomorrow.transform(df['RainTomorrow'])

df = df.dropna()


df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_windgustdir  = le.fit(df['WindGustDir'])
df['WindGustDir'] = df_windgustdir.transform(df['WindGustDir'])

df_winddir9am = le.fit(df['WindDir9am'])
df['WindDir9am'] = df_winddir9am.transform(df['WindDir9am'])

df_winddir3pm = le.fit(df['WindDir3pm'])
df['WindDir3pm'] = df_winddir3pm.transform(df['WindDir3pm'])

df_raintoday = le.fit(df['RainToday'])
df['RainToday'] = df_winddir3pm.transform(df['RainToday'])

df.info()
```

四、进行模型分类预测

```python
X = df.loc[:,df.columns!='RainTomorrow']
y = df.loc[:,df.columns=='RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split

X_tr,X_ts,y_tr,y_ts = train_test_split(X,y)
X_tr.shape,X_ts.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc = dtc.fit(X_tr,y_tr)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dtc,X,y,cv=5,scoring='accuracy')
scores.mean()
```

```python
from sklearn.model_selection import GridSearchCV

parameters={'min_samples_split' : list(range(2,6,1)),
            'max_depth': list(range(6,19,2)),
            'criterion':('gini','entropy'),
            'min_samples_leaf':list(range(2,9,2))}

clf = DecisionTreeClassifier()
gs = GridSearchCV(estimator=clf, param_grid = parameters, scoring='accuracy',iid=False,cv=5,return_train_score=True)
gs.fit(X_tr, y_tr)
gs.best_params_, gs.best_score_
```

```python
dtc_best = gs.best_estimator_
dtc_best.fit(X_tr,y_tr)

y_pred = dtc_best.predict(X_ts)
y_pred[0:5]
```

```python
scores = cross_val_score(dtc_best,X,y,cv=5,scoring='accuracy')
scores.mean()
```

该作业运用决策树优化模型进行分类预测。通过处理缺失值，并将object型变量转化为数值型变量，并用格搜索最终得到精度为0.84的模型。0.84的精度说明基于决策树的优化模型没有出现过拟合或欠拟合现象，模型比较适合此数据。
