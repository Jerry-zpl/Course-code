二、缺失值处理

1、把缺失值太多的数据直接去掉
原因：一共1400多条数据，就确实1300多条，无法进行分析。

```python
df = df.drop(['PoolQC','Fence','MiscFeature','Alley','FireplaceQu'],axis=1)
```

2、缺失值处理：利用逻辑回归的方式进行填补缺失值LotFrontage

```python
df = pd.get_dummies(df)
df_clean = df.dropna()
df_clean_X = df_clean.drop(columns=['SalePrice','LotFrontage','MasVnrArea','GarageYrBlt'])
df_clean_y = df_clean.SalePrice
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso

rfr = RandomForestRegressor()
lasso = Lasso()
```

```python
from sklearn.model_selection import cross_val_score
rfr_cv = cross_val_score(rfr,df_clean_X,df_clean_y,scoring='neg_mean_squared_error',cv=5)
rfr_cv.mean()
```

```python
lasso_cv = cross_val_score(lasso,df_clean_X,df_clean_y,scoring='neg_mean_squared_error',cv=5)
lasso_cv.mean()
```

```python
from sklearn.model_selection import GridSearchCV
param_rfr = {'n_estimators':[10,30,100],
             'max_depth':list(range(2,11,2)),
             'min_samples_split':list(range(2,5,1)),
             'min_samples_leaf':list(range(1,4,1))}
# 使用格搜索查找最优超参数，优化目标是MSE
rfr_gs = GridSearchCV(rfr,param_rfr,scoring='neg_mean_squared_error',cv=5,verbose=1,
                      n_jobs=3).fit(df_clean_X,df_clean_y)
# 查看最优分数、最优超参数
rfr_gs.best_score_,rfr_gs.best_params_
```

```python
# 使用格搜索查找Lasso最优超参数，因为Lasso超参数很少，因此没有单独设置需搜索的超参数字典。
lasso_gs = GridSearchCV(lasso,{'alpha':[1,0.1,0.01]},scoring='neg_mean_squared_error',
                        cv=5).fit(df_clean_X,df_clean_y)
# 查看分数和超参数，比较两个模型分数，选择RF作为Age缺失值的预测函数
lasso_gs.best_score_,lasso_gs.best_params_
```

```python
df_clean_na = df.loc[df.LotFrontage.isna(),:]
df_clean_na_X = df_clean_na.drop(columns=['SalePrice','LotFrontage','MasVnrArea','GarageYrBlt'])
# 预测Age缺失值
df_clean_na_y = rfr_gs.best_estimator_.predict(df_clean_na_X)
# 保存为DataFrame格式（为了将index和数据结合起来，方便转化为dict）
df_clean_na_y = pd.DataFrame(df_clean_na_y,index=df_clean_na_X.index,columns=['LotFrontage'])
df = df.fillna(dict(df_clean_na_y))
df.isna().sum()
```

3、缺失值处理：利用逻辑回归的方式进行填补缺失值GarageYrBlt

```python
mean_garageyrblt = df['GarageYrBlt'].mean()
tmp = df['GarageYrBlt'].copy()
tmp[df.GarageYrBlt.isnull()] = mean_garageyrblt
df['GarageYrBlt_fill'] = tmp
del tmp
df = df.drop(['GarageYrBlt'],axis=1)
```
